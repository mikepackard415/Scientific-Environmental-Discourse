```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
user_agent = 'ozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.81 Safari/537.36'
referer = 'https://www.journals.uchicago.edu/toc/reep/current'
headers = {'User-Agent': user_agent, 'referer':referer}
```

```python
dfs = []
for i, yr in enumerate(range(2007, 2022)):
    for v in [1, 2]:
        vol = 'https://www.journals.uchicago.edu/toc/reep/{}/{}/{}'.format(yr, i+1, v)
        
        response = requests.get(vol, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        url = ['https://www.journals.uchicago.edu'+e.find('a')['href'] 
             for e in soup.find_all('h5', {'class':'issue-item__title'}) if e.text != 'Announcements']
        
        df = pd.DataFrame({'url': url})
        df['volume'] = v
        df['year'] = yr
        dfs.append(df)
```

```python
reep = pd.concat(dfs)
reep.head()
```
